```python
import pandas as pd
import ipdb
```

```python
file = pd.ExcelFile("./Downloads/messy.xlsx")
df = pd.read_excel(file, 'test')
df
print(df.columns)
df = df.drop('Unnamed: 2', axis=1)
```

```python
df.columns = df.columns.to_series().apply(lambda x: x.lstrip().lower().replace('%','').replace(" ","_"))
df
```

```python
df['fll_nam'] = df['fll_nam'].str.title()
df
```

```python
print(df.dtypes)
```

```python
df['temp_split'] = df['fll_nam'].str.split()
df['email'] = ""
for idx, name in enumerate(df['temp_split'].values):
    df.at[idx, 'email'] = name[1].lower() + "." + name[0].lower() + "." + str(df['cust_id'].values[idx]) + "@yourcompany.com"
df = df.drop('temp_split', axis = 1)
df
```

```python
df['mobiles'] = df['mobiles'].astype('string')
print(df.dtypes)
```

```python
def modifyMobile(mobile):
    if mobile[:2] == "84":
        return mobile
    else: return "84" + mobile
df['mobiles'] = df['mobiles'].apply(lambda x: modifyMobile(x))
df
```

```python
df['join_date'] = df['join_date'].astype("string")
def removeFault(date):
    if len(date) > 10: 
        return date[:date.index(" ")]
    else: return date
df['join_date'] = df['join_date'].apply(lambda x: removeFault(x))
df['join_date'] = pd.to_datetime(df['join_date'], infer_datetime_format=True, format='%y-%m-%d')
df
```

```python
df_duplicated_removed = df.sort_values(by=['join_date'])
df_duplicated_removed = df_duplicated_removed.drop_duplicates(subset=['cust_id'], keep="first")
df_duplicated_removed = df_duplicated_removed.sort_values(by=['cust_id'])
```
